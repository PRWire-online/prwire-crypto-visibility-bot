```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/crypto_visibility_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
signal_cols = ['index_tracking_score','algorithmic_visibility_score','entity_authority_score','backlink_quality_score','ai_discovery_score','distribution_reach_score']
signal_avgs = df[signal_cols].mean()
colors = ['#10b981','#3b82f6','#8b5cf6','#f59e0b','#ef4444','#6366f1']
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('PRWire - Average Crypto Visibility Signal Scores', fontsize=14, fontweight='bold')
plt.xlabel('Visibility Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=70, color='green', linestyle='--', alpha=0.5, label='Healthy threshold (70)')
plt.legend()
plt.tight_layout()
plt.show()
print(signal_avgs.round(1).to_string())
```

```python
network_cols = ['google_index','bing_index','chatgpt_discovery','perplexity_ai','google_ai_overviews']
network_avgs = df[network_cols].mean()
plt.figure(figsize=(10,5))
plt.bar(network_avgs.index, network_avgs.values, color=['#10b981','#3b82f6','#f59e0b','#8b5cf6','#6366f1'])
plt.title('PRWire - Average Distribution Network Scores', fontsize=14, fontweight='bold')
plt.xlabel('Distribution Network')
plt.ylabel('Average Score (0-100)')
plt.tight_layout()
plt.show()
```
